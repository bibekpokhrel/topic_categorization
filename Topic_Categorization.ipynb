```python
from pprint import pprint
import pandas as pd
import numpy as np
from json import loads
from collections import defaultdict
```

```python
df = pd.read_json(data)
```

```python
df = df.fillna(0)
df
```

```python
X = df.values
topics = df.columns
```

```python
json_data = loads(data)
json_data
```

```python
occurence = defaultdict((dict))
```

```python
occurence = defaultdict((dict))
for l in json_data:
    for i in l:
        for j in l:
            try:
                occurence[i][j]+= l[j]
            except KeyError:
                occurence[i][j] = l[j]
print(occurence)
```

```python
features = np.matmul(X.T>0,X)
for i in range(len(topics)):
    features[i,:] = features[i,:]/features[i,i]
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
np.fill_diagonal(features,0)
features = StandardScaler().fit_transform(features)
```

```python
clf = KMeans(n_clusters=5, n_init=40)
```

```python
clf.fit(features)
```

```python
category = clf.labels_
category
```

```python
cat = defaultdict(list)
for  i in range(len(topics)):
    cat[category[i]].append(topics[i])
```
